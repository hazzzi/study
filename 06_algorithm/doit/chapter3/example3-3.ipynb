```python
# 실습 3-4
# 이진 검색 알고리즘

from typing import Sequence, Any

def bin_search(a: Sequence, key: Any) -> int:
    """시퀀스 a에서 key와 일치하는 원소를 이진 검색"""
    pl = 0
    pr = len(a) - 1

    while True:
        pc = (pl + pr) // 2
        if a[pc] == key:
            return pc
        elif a[pc] < key:
            pl = pc + 1
        else:
            pr = pc - 1
        if pl > pr:
            break
    return -1

if __name__ == '__main__':
    num = int(input('원소 갯수를 입력'))
    x = [None] * num

    print('배열 데이터를 오름차순으로 정렬')

    x[0] = int(input('x[0] = '))

    for i in range(1, num):
        while True:
            x[i] = int(input(f'x[{i}] = '))
            if x[i] >= x[i-1]:
                break
        
    
    ky = int(input('검색값 입력'))
    idx = bin_search(x, ky)

    if idx == -1:
        print('검색값이 존재하지않습니다')
    else:
        print(f'검색 값은 x[{idx}] 에 있습니다. ')
```
